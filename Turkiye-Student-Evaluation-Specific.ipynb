## About Turkiye-Student-Evaluation-Specific

- attendance: Code of the level of attendance; values from {0, 1, 2, 3, 4}
- difficulty: Level of difficulty of the course as perceived by the student; values taken from {1,2,3,4,5}
- Q1:  The semester course content, teaching method and evaluation system were provided at the start.

## Quels sont les cours enseignÃ©s par chacun des profs ?

```python
courses_by_instructor = {}

for index, row in df.iterrows():
    instructor_id = row['instr']
    course_code = row['class']

    if instructor_id not in courses_by_instructor:
        courses_by_instructor[instructor_id] = []

    courses_by_instructor[instructor_id].append(course_code)

for instructor_id, courses_taught in courses_by_instructor.items():
    print(f"Instructor {instructor_id} teaches courses: {courses_taught}")
```

Selon ces rÃ©sultats, nous avons les cours enseignÃ©s par chaque instructeur :

- Pour l'instructeur 1 : les cours sont 2, 7 et 10.
- Pour l'instructeur 2 : les cours sont 1, 6, 11 et 13.
- Pour l'instructeur 3 : les cours sont 3, 4, 5, 8, 9, 12 et 13.

```python
# Initialiser un dictionnaire pour stocker les cours enseignÃ©s par chaque instructeur
courses_by_instructor = {}

## Initialiser un dictionnaire pour enregistrer le total des rÃ©pÃ©titions par nombre de fois que l'Ã©tudiant suit ce cours.
repeat_by_student = {}

# Initialiser un dictionnaire pour enregistrer le niveau total de difficultÃ© du cours tel que perÃ§u par l'Ã©tudiant.
difficulty_of_the_course = {}

# Initialiser un dictionnaire pour stocker la prÃ©sence moyenne des Ã©tudiants
attendance_by_student = {}

# ItÃ©rer sur les lignes du DataFrame
for index, row in df.iterrows():
    instructor_id = row['instr']
    course_code = row['class']
    repeat = row['nb.repeat']
    difficulty = row['difficulty']
    attendance = row['attendance']

    if instructor_id not in courses_by_instructor:
        courses_by_instructor[instructor_id] = []

    if instructor_id not in repeat_by_student:
        repeat_by_student[instructor_id] = []

    repeat_by_student[instructor_id].append(repeat)

    if instructor_id not in difficulty_of_the_course:
        difficulty_of_the_course[instructor_id] = []

    difficulty_of_the_course[instructor_id].append(difficulty)

    if instructor_id not in attendance_by_student:
        attendance_by_student[instructor_id] = []

    attendance_by_student[instructor_id].append(attendance)

# CrÃ©er un nouveau DataFrame Ã  partir du dictionnaire
result_df = pd.DataFrame(list(courses_by_instructor.items()), columns=['instructor', 'courses'])

# Ajouter la colonne 'average_repeat' au DataFrame rÃ©sultant
result_df['Number of attempts'] = result_df['instructor'].apply(lambda x: sum(repeat_by_student[x]) / len(repeat_by_student[x]))


# Ajouter la colonne 'average_attendance' au DataFrame rÃ©sultant
result_df['average_attendance'] = result_df['instructor'].apply(lambda x: sum(attendance_by_student[x]) / len(attendance_by_student[x]))


# Ajouter la colonne 'average_difficulty' au DataFrame rÃ©sultant
result_df['average_difficulty'] = result_df['instructor'].apply(lambda x: sum(difficulty_of_the_course[x]) / len(difficulty_of_the_course[x]))

# Afficher le DataFrame

Final = pd.DataFrame({'instructor': result_df['instructor'] ,'courses':  [ [2, 7, 10], [ 1, 6, 11 ,13], [3, 4, 5, 8, 9, 12, 13] ], 'Number of attempts':result_df['Number of attempts'], 
                     'average_attendance':result_df['average_attendance'], 'average_difficulty': result_df['average_difficulty'] })
Final
```

```python
n = df_q.shape[0] 
p = df_q.shape[1] 
eigval = (n-1) / n * pca.explained_variance_ # valeurs propres
sqrt_eigval = numpy.sqrt(eigval) # racine carrÃ©e des valeurs propres
corvar = numpy.zeros((p,p)) # matrice vide pour avoir les coordonnÃ©es
for k in range(p):
    corvar[:,k] = pca.components_[k,:] * sqrt_eigval[k]
# on modifie pour avoir un dataframe
coordvar = pd.DataFrame({'id': df_q.columns, 'COR_1': corvar[:,0], 'COR_2': corvar[:,1]})
coordvar
```

Une fois ces coordonnÃ©es rÃ©cupÃ©rÃ©es, on peut donc afficher ce cercle des corrÃ©lations comme ci-dessous.

```python
# CrÃ©ation d'une figure vide (avec des axes entre -1 et 1 + le titre)
fig, axes = plt.subplots(figsize = (13,13))
fig.suptitle("Cercle des corrÃ©lations")
axes.set_xlim(-1, 1)
axes.set_ylim(-1, 1)

# Ajout des axes
axes.axvline(x = 0, color = 'lightgray', linestyle = '--', linewidth = 1)
axes.axhline(y = 0, color = 'lightgray', linestyle = '--', linewidth = 1)

# Ajout des noms des variables
for j in range(p):
    axes.text(coordvar["COR_1"][j],coordvar["COR_2"][j], coordvar["id"][j])
    
# Ajout du cercle
plt.gca().add_artist(plt.Circle((0,0),1,color='blue',fill=False))

plt.show()
```

Dans cette analyse du cercle de corrÃ©lation, nous pouvons interprÃ©ter que l'axe 1 distingue clairement les questions ayant des valeurs Ã©levÃ©es (toutes situÃ©es Ã  gauche) de celles ayant des valeurs faibles (situÃ©es Ã  droite).

## PCA for feature dimensional reduction

The principal component analysis reduces the dimensionality of large data sets by transforming a large group of variables into a smaller one that still contains most of the information in the large set. This method reduces the computational time to train the model and it will also retain the information of the data set.

```python
from sklearn.cluster import KMeans
```

```python
pca=PCA(n_components=2)
dfpca=pca.fit_transform(df_q)
dfpca
```

```python
pca.explained_variance_ratio_.cumsum()[1]
```

```python
clusterrange=range(1,10)
clustererror=[]
for i in clusterrange:
    clusters=KMeans(n_clusters=i,n_init=10,max_iter=100,random_state=42)
    clusters.fit(dfpca)
    clustererror.append(clusters.inertia_)
pd.DataFrame({'number of clusters :':clusterrange, 'Error' : clustererror})
```

## Checking the number of clusters can be made using Elbow method

```python
plt.plot(clusterrange, clustererror )
plt.title('Elbow Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Error')
plt.show()
```

## The Elbow graph suggests that opting for 3 clusters would be appropriate

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
y_kmeans=kmeans.fit_predict(dfpca)
```

```python
plt.scatter(dfpca[y_kmeans == 0, 0], dfpca[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(dfpca[y_kmeans == 1, 0], dfpca[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(dfpca[y_kmeans == 2, 0], dfpca[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
